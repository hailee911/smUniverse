```python
import pandas as pd
import numpy as np
df = pd.read_excel('score.xlsx')
df
```

```python
# 결측치(NaN데이터)
# NaN데이터 채우기 :fillna()
df.fillna('없음',inplace=True)
df
```

```python
# 학교 컬럼에 NaN데이터 입력:
df['학교'] = np.nan
df.fillna('없음',inplace=True)

df['SW특기'].fillna('없음',inplace=True)
df
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.dropna() # NaN 데이터가 있는 row를 모두 삭제

df.dropna(axis='index',how='any')
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
# NaN 4,5번 row 데이터가 삭제
df.dropna(axis='index',how='any')

# axis : 축 - index: row삭제, columns:column 삭제
# how default : any- NaN값이 1개라도 있으면 삭제, all - 전부 NaN일때 삭제됨.
df.dropna(axis='columns')
```

```python
df['학교'] = np.nan
```

```python
df
```

* fillna : NaN 데이터를 해당값으로 채워줌
* dropna : NaN 데이터를 삭제
- axis : 축 indent-row방향으로 삭제, columns-columns 방향으로 삭제
- how : 어떤경우에 삭제할지 결정 any-1개라 NaN값이 있을때 삭제, all- 모두 NaN일때 삭제

- np.nan : numpy라이브러리 NaN입력
