```python
w = 0.25
index = range(len(df))  # 'index'는 'df'의 길이에 따라 설정합니다.

plt.bar(index, df['국어'], width=w, label='국어')
plt.bar([i - w for i in index], df['영어'], width=w, label='영어')
plt.bar([i + w for i in index], df['수학'], width=w, label='수학')
plt.xticks(index, df['이름'], rotation=45)
plt.ylim(0, 130)

# 각 성적값을 그래프에 추가
for i, name in enumerate(df['이름']):
    plt.text(i, df['국어'][i] + 2, str(df['국어'][i]), ha='center', color='blue')
    plt.text(i - w, df['영어'][i] + 2, str(df['영어'][i]), ha='center', color='orange')
    plt.text(i + w, df['수학'][i] + 2, str(df['수학'][i]), ha='center', color='green')

plt.legend(ncol=3, loc='upper center')
plt.show()
```
