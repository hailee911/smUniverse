```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
```

```python
# x축 : 지원번호 , y축 : 국어
# 선색상 : red, marker 색상 : yellow, 테투리 : blue, 선타입 : --
# 범례

# 엑셀 파일 읽기
df = pd.read_excel('score.xlsx', index_col='지원번호')

# 선 그래프 그리기
# plt.plot(df.index, df['국어'], color='red', linestyle='--', marker='o', markerfacecolor='yellow', markeredgecolor='blue', label='국어 점수')

x = df.index
k = df['국어']
e = df['영어']

# 그래프 크기 조정, dpi: 그래프 해상도 설정 facecolor = 배경색
plt.figure(figsize=(20,5),facecolor='purple') # x축으로 5, y축 20 만큼 크기 조절

plt.plot(x,k,label = '국어',marker='o',color='red',ms=15,mfc='yellow',mec='blue',ls='--')
plt.plot(x,e,label = '영어',marker='v',ms=15)

# 그래프 제목과 축 레이블 설정
plt.title('국어 점수 그래프')
plt.xlabel('지원번호')
plt.ylabel('점수')

# 범례 추가
plt.legend()

# 그래프 파일로 저장 dpi:해상도
# plt.savefig('graph.png',dpi=100)
# plt.savefig('graph2.png',dpi=300)


# 그래프 표시
plt.show()


```
