컬럼 이름 전체적으로 한번에 바꾸기

```python
df.columns = ['name','school','height','kor','eng','math','science','society','sw']
df
```

```python
df['school'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)
df['sw'] = df['sw'].str.lower()
df.loc['4번','sw'] = 'django'
df.loc['5번','sw'] = 'python'
df['합계'] = df['kor']+df['eng']+df['math']+df['science']+df['society']
df['결과'] = 0
filt = (df['합계']>=400)
df.loc[filt,'결과'] = 'Pass'
df.loc[~filt,'결과'] = 'Fail'

df
```

타입을 변경해서 문자 추가하는 함수

```python
def add_cm(height):
  return str(height)  + 'cm'
```

```python
df['키'] = df['height'].apply(add_cm)
df
```

함수 호출 .apply(함수이름)

```python
# 함수 선언
def change(lang): # lang타입 : str
  if pd.notnull(lang):
    return lang.upper() # lang.str.uppder() -> list에서 1개 str
  else:
    return 'PYTHON'
```

```python
df['sw'].apply(change)
```
