```python
import pandas as pd
import random
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# 컬럼 전체를 수정
df['요일'] = 5
```

```python
# 특정 셀 수정
df.loc['1번','요일'] = 5
df.iloc[1,9] = 2

# 랜덤으로 0-6까지 숫자 입력
for i in range(8):
  df.iloc[i,9] = random.randint(0,6)
```

```python
df
```

```python
def func(x):
  if x == 0: return '일요일'
  elif x == 1: return '월요일'
  elif x == 2: return '화요일'
  elif x == 3: return '수요일'
  elif x == 4: return '목요일'
  elif x == 5: return '금요일'
  elif x == 6: return '토요일'
  else: return 0
```

```python
df['요일'] = df['요일'].apply(func)
df
```

```python
# 결과 컬럼 생성
df['결과'] = '불합격'

# 국어점수가 60점 이상이면 합격으로 수정하시오.
filt = (df['국어'] >= 60)
df.loc[filt,'결과'] = '합격'
df
```

```python
# 학교+등학교
# 키+'cm' .astype()
df['키']  = df['키'].astype(str)+'cm'
df['학교'] = df['학교']+'등학교'
df
```

```python
df.groupby('학교')['이름'].count()
df.groupby('학교')['이름'].sum()
```

* 컬럼이 숫자가 아니면 실행이 안됨

```python
df.groupby('학교')['이름'].value_counts()
df.groupby('학교')['국어'].mean()
df.groupby('학교')['국어'].count()
df.groupby('학교')['국어'].sum().sort_values(ascending=False)
```

```python
df = pd.read_csv('drinks.csv')
df.info()
```

```python
# 모든 술의 총량을 출력
# total_servings 컬럼을 구하시오. beer + spirit + wine
# / 전체 알콜 도수를 확인
df['total_servings'] = df['beer_servings']+df['spirit_servings']+df['wine_servings']

df
```

```python
df['continent'].unique()
```

```python
filt = df['continent'].isnull()
df[filt]['total_servings']
df.loc[filt,'total_servings']
df[filt]['total_servings'].sum()
# 이름이 없는 대륙의 총 서빙량
```

```python
# 대륙 NaN 값 -> OT 로 대체
df['continent']= df['continent'].fillna('OT')
```

```python
# 대륙별 total_servings를 구하시오
df.groupby('continent')['total_servings'].sum().sort_values()
```

```python
df.columns
```

```python
# 술 소비량 대비 알콜 비율이 어느 정도 되는지 출력
# 알콜 비율을 계산해서 컬럼을 추가
# total_litres_of_pure_alcohol 알콜 비율을 total_servings 으로 나눈것
# 술 소비량 대비 알콜 비율이 높은 나라별 20개를 출력하시오.

df['alcohol_rate'] = df['total_litres_of_pure_alcohol'] / df['total_servings']
df['alcohol_rate'] = df['alcohol_rate'].fillna(0)
df.groupby('country')['alcohol_rate'].sum().sort_values(ascending=False).head(20)

# 대륙별 알콜비율도 계산해서 출력하시오.
df.groupby('continent')['alcohol_rate'].sum()

df[['country','alcohol_rate']].head(20).sort_values('alcohol_rate',ascending=False)
```
