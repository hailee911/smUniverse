```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
```

```python
x = [1,2,3]
y = [2,4,8] # SM 매출액
y2 = [5,1,3] # LG 매출액
y3 = [1,2,5] # SK 매출액

plt.figure(figsize=(10,5))
plt.plot(x,y,label='SM매출액', marker='o', ls='--')
plt.plot(x,y2,label='LG매출액', marker='x', ls='-.')
plt.plot(x,y3,label='SK매출액', marker='v', ls=':')

# for idx,txt in enumerate(y):
#   plt.text(x[idx],y[idx]+0.2, txt, ha='center')

for i in range(3):
  plt.text(x[i],y[i]+0.2,y[i], ha='center')
  plt.text(x[i],y2[i]+0.2,y[i], ha='center')
  plt.text(x[i],y3[i]+0.2,y[i], ha='center')

plt.title('기업별 매출액')
plt.legend(ncol=3) # 범례출력 세로출력, ncol = 3
plt.show()
```
