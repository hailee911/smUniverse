```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호
```

```python
# x,y 축의 데이터
x = [1,2,3] # 리스트 튜플 DataFrame 데이터 (리스트 형태)
y = [2,4,8]
yy = [1,8,3]
# linewidth : 선두께
plt.plot(x,y, label='인원변동', linewidth=5 , marker = '*', markersize = 20, markeredgecolor = 'red')
# marker : 지점표시 - o x v * markersize : 마커크기, markeredgecolor = 마커 테두리선 색 , markerfacecolor = 마커 색 채우기 alpha = 투명도
# ls = : 라인 bv: = b 선색상 v 마커표시 :선모양
plt.plot(x,yy, label='비용변동', marker='*',markersize = 20, markerfacecolor = 'yellow', alpha = 0.5, ls=':')
plt.title('라인그래프 Line',size=30)

# 범례 : 그래프 선 설명문구, loc : 위치 조정 upper center, right, left / lower c,r,l
# plt.legend(loc='lower right')
plt.legend(loc=(0.4,0.1)) # 좌표값으로 위치 설정 가능
```
