```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
```

```python
# x,y 축의 데이터
x = [1,2,3] # 리스트 튜플 DataFrame 데이터 (리스트 형태)
y = [2,4,8]

plt.figure(figsize=(7,5))
plt.plot(x,y,label='범례데이터', marker='o')

# text(x좌표위치, y좌표위치, 데이터값)
# plt.text(x[0],y[0],2)

# 각 점에 숫자 표시하기
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', ha='center', va='top', fontsize=15)

plt.title('라인그래프')
plt.legend()
plt.show()
```
