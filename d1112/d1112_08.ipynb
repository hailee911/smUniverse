```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
```

라벨변경 라벨의 기울기 설정 width alpha 투명도 조정

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]

# color width
# plt.bar(label, values, color = 'r') # 1개의 색상으로 처리

colors = ['r','b','g']
plt.figure(figsize=(5,3))

plt.title('막대그래프')
plt.bar(label, values, color = colors, width=0.2, label='학생키')

plt.xticks(label,['1번','2번','3번'],rotation=45)
plt.yticks([0,50,100,150,200,250,300])

plt.legend(loc='upper center')

plt.show()
```

막대 그래프의 무늬 설정

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors ,alpha = 0.5)
# 막대그래프의 무늬설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

* 0부터 범위 설정 xticks yticks - 0부터 표시됨
* 특정구간부터 범위설정 : xlim, ylim

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors ,alpha = 0.5)
# plt.yticks([0,100,150,200]) # 0부터 범위지정
plt.ylim(175,199) # 특정구간 범위 지정
plt.show()
```

```python
label = ['홍길동','유관순','강감찬']
colors = ['red','green','blue']
values = [180,187,184]
plt.barh(label,values, height=0.5, color = colors)
# 특정구간 범위 지정
plt.xlim(175,190)
```
