문자열 함수 : startwith, contains, replace

```python
# 해당 값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')

# 해당 값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt,['이름','학교','키']]
```

```python
langs = ['python','java']
filt = df['SW특기'].isin(langs)

# 해당값을 모두 소문자로 변환
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# # 포함이 되어있으면 모두 검색
# -nan 데이터 처리를 하지 않으면 에러
filt = df['SW특기'].str.contains('Java',na=False)
df[filt]

# isin() ; 관호 안에는 리스트 타입만 가능 str 불가
langs = ['p','j']

# 컬럼 수정 방법
df['SW특기'] = df['SW특기'].str.lower()
```

```python
filt = df['SW특기'].str.lower().isin({'Python','Java'})
df[filt]
```

```python
# j 가 포함되어 있는 정보 검색
filt = df['SW특기'].str.contains('j',na=False)
df[filt]
```

널값 처리 : fillna dropna

```python
import numpy as np
```

```python
df.fillna('데이터 없음',inplace=True)
df
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df['학교'] = np.nan # 학교 컬럼에 NaN데이터 입력
df.fillna('데이터 없음')
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# dropna
# axis 축방향으로 삭제 : index(row방향),columns(columns방향)
# how : any (1개라도 있으면 모두row나columns삭제), all(모두 nan데이터일 때 삭제)
# axis = 'index','columns','how = 'any','all'
df.dropna() # index,any 기본 default - 해당되는 index(row)를 삭제

df.dropna(axis='index',how='all') # 가로줄에 모두 난이면 삭제 -> 없으므로 삭제 안됨.

df.dropna(axis='columns',how='all') # 학교 컬럼이 모두 난이기에 삭제됨

df.dropna(axis='index',how='any') # 가로줄에 학교컬럼이 nan이 있기에 모두 삭제됨

df.dropna(axis='columns',how='any')
```
