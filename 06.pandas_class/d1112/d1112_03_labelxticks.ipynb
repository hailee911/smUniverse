```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호
```

```python
# x,y 축의 데이터
x = [1,2,3] # 리스트 튜플 DataFrame 데이터 (리스트 형태)
y = [2,4,8]
plt.plot(x,y)
plt.title('라인그래프 Line')

# 그래프 축 라벨, 글자색 변경

# 그래프 X축 설명, color 글자 색 변경 loc 글자 위치
plt.xlabel('x축 날짜', color='#00ff00',loc='center') # 그래프 X축 설명
plt.ylabel('Y축 인원', color ='blue', loc='center') # 그래프 Y축 설명

plt.xticks([1,2,3])
plt.yticks([2,3,4,8,10])

plt.show()
```
