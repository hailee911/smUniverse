```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
```

```python
df = pd.read_csv('drinks.csv')

df['country'].index
```

drink.csv 파일 불러와서

beer_servings,spirit_servings,wine_servings 합계를

total_servings 컬럼을 생성해 추가하고

country 별 total_servings 막대 그래프를 생성하시오.


```python

# total_servings 컬럼 추가
df['total_servings'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']


# 막대 그래프 생성
plt.figure(figsize=(10, 8))
plt.bar(df['country_code'], df['total_servings'], color='skyblue')
plt.xlabel('Country Code')
plt.ylabel('Total Servings')
plt.title('Total Servings of Alcohol by Country')
plt.show()
```
