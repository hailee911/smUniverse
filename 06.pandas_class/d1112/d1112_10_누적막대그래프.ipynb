```python
df = pd.read_excel('score.xlsx')
```

* 누적막대그래프
* bottom : 첫번째 값을 입력

```python
plt.figure(figsize=(7,4))
plt.bar(df['이름'], df['국어'], label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], label='수학')
plt.xticks(rotation=45)
plt.legend(ncol=3, loc='upper center')
plt.ylim(0, 350)

# 각 성적값을 그래프에 추가
for i, name in enumerate(df['이름']):
  # 국어 점수 위치
  plt.text(i, df['국어'][i] / 2, str(df['국어'][i]), ha='center', va='center', color='white')
  
  # 영어 점수 위치
  plt.text(i, df['국어'][i] + df['영어'][i] / 2, str(df['영어'][i]), ha='center', va='center', color='white')
  
  # 수학 점수 위치
  plt.text(i, df['국어'][i] + df['영어'][i] + df['수학'][i] / 2, str(df['수학'][i]), ha='center', va='center', color='white')

plt.show()
```
