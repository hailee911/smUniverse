```python
import pandas as pd
```

```python
# DataFrame 변경해서
# 번호 이름 국어 영어 수학 합계 평균 등수
# 번호 이름 합계 평균 출력
```

```python
data = {
  '번호': [1, 2, 3, 4, 5, 6, 7],
  '이름': ['홍길동', '유관순', '이순신', '강감찬', '김구', '김유신', '홍길자'],
  '국어': [100, 80, 90, 60, 100, 80, 100],
  '영어': [100, 80, 90, 65, 100, 80, 100],
  '수학': [99, 85, 91, 67, 84, 71, 90],
  '합계': [299, 245, 271, 192, 284, 231, 290],
  '평균': [99.67, 81.67, 90.33, 64.0, 94.67, 77.0, 96.67],
  '등수': [0, 0, 0, 0, 0, 0, 0]
}
```
