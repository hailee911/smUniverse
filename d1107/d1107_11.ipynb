```python
# 문자열(str) 조건]
# 문자열함수 endwith strip replace
df = pd.read_excel('score.xlsx')

filt = df['이름'].str.startswith('강')
df[filt]

filt = df['이름'].str.contains('김')
df[~filt]
```

```python
# 이름 t 가 들어가면서 평균 80점 이상 출력
df['SW특기']
langs = ['Python','Java']
filt = df['SW특기'].isin(langs)
df[filt]

# 소문자 치환 : isin() 확인
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
# Java라는 글자가 있는지 체크
df['SW특기'].str.contains('Java', na=False)


# na = True : NaN True 인식, na = False : NaN False 인식
```
