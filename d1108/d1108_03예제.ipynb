```python
import pandas as pd
import numpy as np
```

컬럼 2개 구성 (가중치1,가중치2) 
전체 가중치1 (사망+부상) * 1.1

2014 사망 부상 가중치2 (사망+부상) *1.2

2015 사망 부상 가중치2 (사망+부상) *1.5

2016 사망 부상 가중치2 (사망+부상) *1.3

```python
df1 = pd.read_csv('2014년졸음운전교통사고.csv',encoding='euc-kr')
# 조건식 2014년에 해당되는 조건식 구성
filt = df1['구분'].str.slice(start=0,stop=4) =='2014'
```

```python

df1['가중치1'] = (df1['사망(명)']+df1['부상(명)'])*1.1
df1['가중치2'] = (df1['사망(명)']+df1['부상(명)'])*1.2
df1
```

```python
df2 = pd.read_csv('2015년졸음운전교통사고.csv',encoding='euc-kr')
filt = df2['구분'].str.slice(start=0,stop=4) =='2015'
df2['가중치1'] = (df2['사망(명)']+df2['부상(명)'])*1.1
df2['가중치2'] = (df2['사망(명)']+df2['부상(명)'])*1.5
df2
```

```python
df3 = pd.read_csv('2016년졸음운전교통사고.csv',encoding='euc-kr')
filt = df3['구분'].str.slice(start=0,stop=4) =='2016'
df3['가중치1'] = (df3['사망(명)']+df3['부상(명)'])*1.1
df3['가중치2'] = (df3['사망(명)']+df3['부상(명)'])*1.3
df3
```

```python
df = pd.concat([df1,df2,df3])
df
```
