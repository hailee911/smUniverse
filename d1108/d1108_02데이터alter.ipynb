```python
import pandas as pd
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

데이터 수정

```python
# df['학교'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)
df
```

```python
df['SW특기'].str.lower() # 소문자

# 특정 컬럼 값 변경
df['SW특기'] = df['SW특기'].str.upper() #대문자
df['SW특기'] = df['SW특기'].str.capitalize() #첫글자 대문자
df['키'] = 200
```

```python
df.loc[['2번','3번'],['국어','영어']] = 50
```

```python
df['학교'] = df['학교']+'등학교'
df
```

```python
df.info()
```

데이터 추가

```python
# 데이터 추가
# 컬럼에 없는 컬럼 이름을 입력하면 컬럼이 추가 됨
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계']/5
```

조건식을 사용해서 값 변경

```python
# 결과 컬럼 추가
df['결과'] = 'Fail'

# 결과 컬럼에서 -> 조건식을 사용해서 값을 변경
filt = (df['합계'] >= 400)
df.loc[filt,'결과'] = 'Pass'
```

```python
df['합산'] = 100
# 국어 점수가 90점 이상이면 df['합산'] + 100

filt = (df['국어'] >= 90)
df.loc[filt,'합산'] = df.loc[filt,'합산'] + 100

df
```
